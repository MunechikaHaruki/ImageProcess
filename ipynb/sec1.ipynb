```python
import cv2
import matplotlib.pyplot as plt

print(cv2.__version__)
img = cv2.imread("./image.png")
```

```python
resize_image = cv2.resize(img, (100, 100))  # 640x480にリサイズ
plt.imshow(cv2.cvtColor(resize_image, cv2.COLOR_BGR2RGB))
```

```python
def rotate_image(image, angle):
    # 画像の中心を取得
    height, width = image.shape[:2]
    center = (width / 2, height / 2)

    # 回転行列を作成
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

    # アフィン変換を実行
    rotated = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated


# 45度回転の例
angle = 45  # 任意の角度を指定（反時計回りが正）
rotated_img = rotate_image(img, angle)

plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
```

```python
def mirror_image(image, flip_code):
    """
    画像を反転させる関数
    flip_code:
        1  -> 水平方向の反転（左右反転）
        0  -> 垂直方向の反転（上下反転）
        -1 -> 両方向の反転（上下左右反転）
    """
    return cv2.flip(image, flip_code)


# 使用例
# 左右反転
horizontal_flip = mirror_image(img, 1)
(
    plt.subplot(221),
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),
    plt.title("Original"),
)
(
    plt.subplot(222),
    plt.imshow(cv2.cvtColor(horizontal_flip, cv2.COLOR_BGR2RGB)),
    plt.title("y軸"),
)

# 上下反転
vertical_flip = mirror_image(img, 0)
(
    plt.subplot(223),
    plt.imshow(cv2.cvtColor(vertical_flip, cv2.COLOR_BGR2RGB)),
    plt.title("x軸"),
)

# 上下左右反転
both_flip = mirror_image(img, -1)
(
    plt.subplot(224),
    plt.imshow(cv2.cvtColor(both_flip, cv2.COLOR_BGR2RGB)),
    plt.title("直線y=x"),
)

plt.tight_layout()
plt.show()
```
