```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 
img_path = "image.png"  #  
img = cv2.imread(img_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
def translate_image(image, tx=0, ty=0):
    """
    

    Parameters:
        image: 
        tx: x
        ty: y
    """
    rows, cols = image.shape[:2]

    # 
    transform_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

    # 
    translated = cv2.warpAffine(image, transform_matrix, (cols, rows))
    return translated


# 
plt.figure(figsize=(15, 10))

# 
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("")

# 
translated_diagonal = translate_image(img, tx=100, ty=100)
plt.subplot(122)
plt.imshow(cv2.cvtColor(translated_diagonal, cv2.COLOR_BGR2RGB))
plt.title("100")

plt.tight_layout()
plt.show()
```

```python
def affine_transform(image, angle=0, scale=1.0, tx=0, ty=0):
    """
    

    Parameters:
        image: 
        angle: 
        scale: 
        tx: x
        ty: y
    """
    rows, cols = image.shape[:2]
    center = (cols / 2, rows / 2)

    # 
    M_rot = cv2.getRotationMatrix2D(center, angle, scale)

    # 
    M_rot[0, 2] += tx
    M_rot[1, 2] += ty

    # 
    transformed = cv2.warpAffine(image, M_rot, (cols, rows))
    return transformed


# 
plt.figure(figsize=(15, 10))

# 
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("")

# 
transformed3 = affine_transform(img, angle=40, scale=0.3, tx=50, ty=50)
plt.subplot(122)
plt.imshow(cv2.cvtColor(transformed3, cv2.COLOR_BGR2RGB))
plt.title(" 30 +  1.5 + (50,50)")

plt.tight_layout()
plt.show()
```

```python
def perspective_transform(image, points_src, points_dst):
    """
    

    Parameters:
        image: 
        points_src: 4
        points_dst: 4
    """
    # 
    matrix = cv2.getPerspectiveTransform(points_src, points_dst)

    # 
    height, width = image.shape[:2]
    warped = cv2.warpPerspective(image, matrix, (width, height))
    return warped


# 
plt.figure(figsize=(15, 10))

# 
height, width = img.shape[:2]

# 4
offset = 100
src_points = np.float32(
    [
        [offset, offset],  # 
        [width - offset, offset],  # 
        [offset, height - offset],  # 
        [width - offset, height - offset],  # 
    ]
)

# 4
dst_points = np.float32(
    [
        [offset * 0.9, offset],  # 
        [width - offset * 0.9, offset],  # 
        [offset, height - offset],  # 
        [width - offset, height - offset],  # 
    ]
)

# 
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("")
# 
plt.plot(src_points[:, 0], src_points[:, 1], "r.")

# 
warped_img = perspective_transform(img, src_points, dst_points)
plt.subplot(122)
plt.imshow(cv2.cvtColor(warped_img, cv2.COLOR_BGR2RGB))
plt.title("")
# 
plt.plot(dst_points[:, 0], dst_points[:, 1], "r.")

plt.tight_layout()
plt.show()
```
